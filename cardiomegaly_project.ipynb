```python
### all the ml libraries imported needed
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
### all the dl libraries imported

import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras import layers
from keras.layers import Dense,Conv2D,MaxPool2D,Flatten,BatchNormalization,Dropout
from keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping
```

```python
train_path='/content/train/train'
filepath=[] ### for storing the file path of each image
labels=[] ### for storing the label of each image

folds=os.listdir(train_path) ### lists all the direc present inside the path true,false
for fold in folds: ### mtlb pehle true waali list mein jayega phir false waali mein aise(loops over each class)

  f_path=os.path.join(train_path,fold)  ###(e.g., /content/train/train/true).
  filelists=os.listdir(f_path)  ### lists all the images in the current file directory
  for file in filelists:
    filepath.append(os.path.join(f_path,file))
    labels.append(fold)

fseries=pd.Series(filepath,name='filepaths')
lseries=pd.Series(labels,name='labels')
train_df=pd.concat([fseries,lseries],axis=1)
```

```python
test_path='/content/test/test'

filepaths=[]
labels=[]

folds=os.listdir(test_path)

for fold in folds:
  f_path=os.path.join(test_path,fold)
  filelists=os.listdir(f_path)
  for file in filelists:
    filepaths.append(os.path.join(f_path,file))
    labels.append(fold)

fseries=pd.Series(filepaths,name='filepaths')
lseries=pd.Series(labels,name='labels')
test_df=pd.concat([fseries,lseries],axis=1)
```

```python
valid,test=train_test_split(test_df,test_size=0.5,random_state=42)
```
